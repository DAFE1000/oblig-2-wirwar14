```python
import numpy as np
import matplotlib.pyplot as plt
 

def f(x):
    return np.exp(-x/4) * np.arctan(x)
 
def g(x):
    # nullpunktet til (f'(x) = 0)
    return np.arctan(x) - 4/(x**2 + 1)
 
def g_prime(x):
    # Derivert 
    return 1/(x**2 + 1) + 8*x/(x**2 + 1)**2
 

x = 2.0   # start
toleranse = 1e-10
 
while True:
    x_ny = x - g(x) / g_prime(x)
    if abs(x_ny - x) < toleranse:
        break
    x = x_ny
 
x_topp = x_ny
y_topp = f(x_topp)
 
print(f"Toppunkt: x = {x_topp:.10f}, y = {y_topp:.10f}")
 
x_verdier = np.linspace(0, 5, 1000)
y_verdier = f(x_verdier)
 
plt.figure(figsize=(9, 5))
plt.plot(x_verdier, y_verdier, color='steelblue', linewidth=2,
         label=r'$f(x) = e^{-x/4} \cdot \arctan x$')
plt.scatter(x_topp, y_topp, c= 'green',
 label=f'Toppunkt ({x_topp:.4f}, {y_topp:.4f})')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'Graf av $f(x) = e^{-x/4} \cdot \arctan x$')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
